```python
import matplotlib.pyplot as plt
from dataclasses import dataclass
```

```python
LABELS = ["all", "caption", "footnote", "formula", "list-item", "page-footer", "page-header", "picture", "section-header", "table", "text", "title"]
LABEL2ID = {k: v for v, k in enumerate(LABELS)}

@dataclass
class Metrics:
    model_size: str
    train_iter: int
    box_precision: list[float]
    recall: list[float]
    map50: list[float]
    map50_95: list[float]

    def __getitem__(self, key):
        """ Return the metrics for a given label key."""
        if key in LABEL2ID:
            idx = LABEL2ID[key]
            return [
                self.box_precision[idx],
                self.recall[idx],
                self.map50[idx],
                self.map50_95[idx],
            ]
    
    def get_metric(self, metric_name: str):
        """ Return the metric values for all labels."""
        if metric_name == "box_precision":
            return self.box_precision
        elif metric_name == "recall":
            return self.recall
        elif metric_name == "map50":
            return self.map50
        elif metric_name == "map50_95":
            return self.map50_95
        else:
            raise ValueError(f"Unknown metric name: {metric_name}")
        
    @staticmethod
    def available_metrics():
        return ["box_precision", "recall", "map50", "map50_95"]
```

```python
def plot_metric(metric_name: str, metrics: list[Metrics], highlight: Metrics = None):
    plt.figure(figsize=(10, 6))
    for metric in metrics:
        plt.plot(LABELS, metric.get_metric(metric_name), marker='o', label=f'Iter {metric.train_iter} ({metric.model_size})', alpha=0.7)
    if highlight:
        plt.plot(
            LABELS,
            highlight.get_metric(metric_name),
            marker='o',
            label=f'Highlighted {highlight.model_size}{highlight.train_iter}',
            color='red',
            linewidth=2.5
        )

    plt.xlabel('Labels')
    plt.ylabel(metric_name)
    plt.title(f'{metric_name} for Different Training Iterations and Model Sizes')
    plt.xticks(rotation=45)
    plt.legend()
    plt.grid()
    plt.tight_layout()
    plt.show()


for m in Metrics.available_metrics():
    plot_metric(m, [train_4, train_9], train_9 )
```

```python
# write the code to plot a bar chart comparing the different metric scores,
# for each category there should be n bars representing the different models
# the x axis are the categories, the y axis is the metric score, and the bars are colored differently for each model
def plot_metric_bars(metric_name: str, metrics: list[Metrics]):
    import numpy as np

    num_labels = len(LABELS)
    num_metrics = len(metrics)

    # Get all scores into a 2D array: rows are models, columns are labels
    all_scores = np.array([m.get_metric(metric_name) for m in metrics])

    # Find the minimum score for each label (column-wise minimum)
    worst_scores = np.min(all_scores, axis=0)

    # Calculate percentage improvement over the worst score for each category
    # Using np.divide to handle potential division by zero safely
    improvement_scores = np.divide(
        (all_scores - worst_scores), 
        worst_scores, 
        out=np.zeros_like(all_scores, dtype=float), 
        where=worst_scores!=0
    ) * 100

    x = np.arange(num_labels)  # the label locations
    width = 0.8 / num_metrics  # the width of the bars, adjusted for number of metrics
    
    fig, ax = plt.subplots(figsize=(14, 8))

    for i, metric in enumerate(metrics):
        offset = width * i - (width * (num_metrics - 1) / 2)
        rects = ax.bar(x + offset, improvement_scores[i], width, label=f'Iter {metric.train_iter} ({metric.model_size})')

    ax.set_xlabel('Labels')
    ax.set_ylabel('Percentage Improvement (%)')
    ax.set_title(f'Percentage Improvement for {metric_name} vs. Worst Performer')
    ax.set_xticks(x)
    ax.set_xticklabels(LABELS, rotation=45, ha="right")
    ax.legend()
    ax.grid(axis='y', linestyle='--', alpha=0.7)
    
    fig.tight_layout()
    plt.show()

for m in Metrics.available_metrics():
    plot_metric_bars(m, [train_4, train_9])

```
